```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("survey.csv")
```

```python
df.head()
```

```python
df.columns = ['date', 'age', 'identity', 'industry', 'state', 'salary', 'statues', 'discriminated', 'location', 'solution','follow', 'contact', 'moved']
```

```python
df.head()
```

```python
df['identity'] = df['identity'].str.split(',')
```

```python
# convert list of pd.Series then stack it
df_id = (df
 .set_index(['date','age','industry','state','salary', 'statues', 'discriminated', 'location', 'solution','follow', 'contact','moved'])['identity']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_12', axis=1)
 .rename(columns={0:'identity'}))
```

```python
df_id.head(27)
```

```python
df = df.drop(columns = ['follow', 'contact'])
```

```python
df.head(50)
```

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
g = sns.barplot(data=df, x = df.discriminated.value_counts().index, y=df.discriminated.value_counts(), palette= 'crest', ci=None)
g.set_xticklabels(g.get_xticklabels(), fontsize=16)



plt.xlabel("Discriminated Against?", fontsize=24)
plt.ylabel("Amount", fontsize=20)
plt.title("Discrimination Ratios", fontsize=32)
plt.tight_layout
plt.show()
```

```python
desc = df[df.discriminated.str.contains("Yes")]
desc_count = desc.discriminated.count()
total = df.discriminated.count()

ratio = (desc_count / total) * 100

print(ratio,"%")
```

```python
df_1 = df[~df.discriminated.str.contains("No")]
```

```python
df_id_2 = df_id[~df_id.discriminated.str.contains("No")]
```

```python
fig, ax = plt.subplots(figsize=(20,8))
g = sns.barplot(data=df_id_2, x = df_id_2.identity.value_counts().index, y=df_id_2.identity.value_counts(), palette= 'crest', ci=None)
g.set_xticklabels(g.get_xticklabels(), rotation=45, ha='right', fontsize=14)

plt.xlabel("Identity", fontsize=20)
plt.ylabel("Amount", fontsize=20)
plt.title("Those Discriminated Against Identfiy As:", fontsize=32)
plt.tight_layout
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
g = sns.barplot(data=df_1, x = df.location.value_counts().index, y=df.location.value_counts(), palette= 'crest', ci=None)
g.set_xticklabels(g.get_xticklabels(),fontsize=16)

plt.xlabel("Yes or No", fontsize=20)
plt.ylabel("Amount", fontsize=20)
plt.title("Those Discriminated in the Workplace:", fontsize=32)
plt.tight_layout
plt.show()
```

```python
df_sal = df_1[df_1['salary'] != 0]
```

```python
fig, ax = plt.subplots(figsize=(20,8))
g = sns.barplot(data=df_sal, x = df_sal.salary.value_counts().index, y=df_sal.salary.value_counts(), palette= 'BuGn')
g.set_xticklabels(['<60k', '61-70k', '71-80k', '81-90k', '91-100k', '101k<'], fontsize=16)

plt.xlabel("Salaries", fontsize=22)
plt.ylabel("Amount", fontsize=20)
plt.suptitle("Distribution of Salaries", fontsize=32)
plt.title("Among People Experiencing Discrimination", fontsize=18)
plt.tight_layout
plt.show()
```

```python
sal = df_sal[df_sal['salary'] >= 100]
sal_count = sal.salary.count()
total_sal = df.salary.count()

ratio_sal = (sal_count / total_sal) * 100

print(ratio_sal,"% Made over 100k a year")

sal = df_sal[df_sal['salary'] <= 60]
sal_count = sal.salary.count()
total_sal = df.salary.count()

ratio_sal = (sal_count / total_sal) * 100

print(ratio_sal,"% Made under 60k a year")

sal = df_sal[(df_sal['salary'] >= 60) & (df_sal['salary'] <= 100)]
sal_count = sal.salary.count()
total_sal = df.salary.count()

ratio_sal = (sal_count / total_sal) * 100

print(ratio_sal,"% Made between 60k and 100k a year")
```

```python
wrk = df_1[df_1.location.str.contains("In workplaces")]
wrk_count = wrk.location.count()
total_wrk = df_1.location.count()

wrk_ratio = (wrk_count / total_wrk) * 100

print(wrk_ratio,"%")
```

```python
solu = pd.read_csv('solutions.csv')
```

```python
solu.columns = ['solutions','id']
```

```python
solu['solutions'] = solu['solutions'].str.split(',')
```

```python
df_sol = (solu
 .set_index(['id'])['solutions']
 .apply(pd.Series)
 .stack()
 .reset_index()
 .drop('level_1', axis=1)
 .rename(columns={0:'solutions'}))
```

```python
df_sol.head(50)
```
